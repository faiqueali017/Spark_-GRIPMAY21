## Toss decision:

## Let us see the toss decisions taken so far.

```python
dset_series = dset_matches.toss_decision.value_counts()
labels = (np.array(dset_series.index))
sizes = (np.array((dset_series / dset_series.sum())*100))
colors = ['gold', 'lightskyblue']
plt.pie(sizes , labels = labels , colors = colors , autopct = "% 1.1f%%",
       shadow = True , startangle=90)
plt.title("Toss Decision percentage")
plt.show()
```

##### Almost 55% of the toss decisions are made to field first. Now let us see how this decision varied over time.

```python
plt.figure(figsize=(12,6))
sns.countplot(x = "season" , hue = "toss_decision" , data = dset_matches)
plt.xticks(rotation = "vertical")
plt.show()
```

##### It seems during the initial years, teams wanted to bat first. Voila.! Look at the 2016 season, most of the toss decisions are to field first

```python
num_of_wins = (dset_matches.win_by_wickets>0).sum()
num_of_loss = (dset_matches.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins / total) *100 , (num_of_loss/total)*100]
colors = ['gold' , 'lightskyblue']
plt.pie(sizes , labels = labels , colors = colors , autopct = "%1.1f%%",
       shadow = True , startangle = 90)
plt.title("Win percentage batting Sceond")
plt.show()
```

### Top players of the match:

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2. , 1.02*height, "%d" % int(height),
               ha="center" , va = "bottom")
```

```python
dset_series = dset_matches.player_of_match.value_counts()[:10]
labels = np.array(dset_series.index)
ind = np.arange(len(labels))
width = 0.9
fig ,ax = plt.subplots(figsize = (15,8))
rects = ax.bar(ind, np.array(dset_series), width= width)
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels , rotation = "vertical")
ax.set_ylabel("Count")
ax.set_title("Top player of the month")
autolabel(rects)
plt.show()
```

### Batsman analysis:

#### Let us start our analysis with batsman. Let us first see the ones with most number of IPL runs under their belt.

```python
temp_dset = dset_score.groupby("batsman")["batsman_runs"].agg("sum").reset_index().sort_values(by = "batsman_runs" , ascending = False).reset_index(drop=True)
temp_dset = temp_dset.iloc[:10,:]
labels = np.array(temp_dset["batsman"])
ind = np.arange(len(labels))
width = 0.9
fig , ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind,np.array(temp_dset["batsman_runs"]),width=width , color = "blue")
ax.set_xticks(ind + ((width)/2.))
ax.set_xticklabels(labels , rotation = "vertical")
ax.set_ylabel("Count")
ax.set_title("Top run scores in IPL")
ax.set_xlabel("Batsman Name")
autolabel(rects)
plt.show()
```

##### Virat Kohli is leading the chart followed closely by Raina. Gayle is the top scorer among foreign players.

```python
temp_dset = dset_score.groupby("batsman")["batsman_runs"].agg(lambda x: (x==4).sum()).reset_index().sort_values(by = "batsman_runs" , ascending = False).reset_index(drop = True)
temp_dset = temp_dset.iloc[:10,:]
labels = np.array(temp_dset["batsman"])
ind = np.arange(len(labels))
width = 0.9
fig ,ax = plt.subplots(figsize = (15,8))
rects = ax.bar(ind, np.array(temp_dset["batsman_runs"]), width = width , color = "lightskyblue")
ax.set_xticks(ind+(width/2.))
ax.set_xticklabels(labels , rotation = "vertical")
ax.set_ylabel("Count")
ax.set_title("Batsman with ost number of boundries : ", fontsize = 10)
autolabel(rects)
plt.show()
```

##### Gambhir is way ahead of others - almost 60 boundaries more than Kohli.! Nice to Sachin in the top 10 list :)

```python
temp_dset = dset_score.groupby("batsman")["batsman_runs"].agg(lambda x: (x==6).sum()).reset_index().sort_values(by = "batsman_runs", ascending = False).reset_index(drop = True)
temp_dset = temp_dset.iloc[:10,:]
labels = np.array(temp_dset["batsman"])
ind = np.arange(len(labels))
width = 0.9
fig , ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind,np.array(temp_dset["batsman_runs"]), width = width , color = "m")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels,rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Batsman with the number of sixes.!")
ax.set_xlabel("Batsman Name")
autolabel(rects)
plt.show()
```

##### There you see the big man. Gayle, the unassailable leader in the number of sixes.

##### Raina is third in both number of 4's and 6's

```python
temp_dset = dset_score.groupby("batsman")["batsman_runs"].agg(lambda x: (x==0).sum()).reset_index().sort_values(by = "batsman_runs", ascending = False).reset_index(drop = True)
temp_dset = temp_dset.iloc[:10,:]
labels = np.array(temp_dset["batsman"])
ind = np.arange(len(labels))
width = 0.9
fig , ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind,np.array(temp_dset["batsman_runs"]), width = width , color = "c")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels,rotation = "vertical")
ax.set_ylabel("Count")
ax.set_title("Batsman with the number of dot Balls!")
ax.set_xlabel("Batsman Name")
autolabel(rects)
plt.show()
```

##### It is interesting to see that the same names repeat again here as well. I think since these guys have played more number of balls, they have more dot balls as well.

```python
def balls_faced(x):
    return len(x)

def dot_balls(x):
    return (x==0).sum()

dset_temp = dset_score.groupby("batsman")["batsman_runs"].agg([balls_faced , dot_balls]).reset_index()
dset_temp = dset_temp.loc[dset_temp.balls_faced>200,:]
dset_temp['percentage_of_dot_balls'] = (dset_temp["dot_balls"] / dset_temp["balls_faced"])*100.
dset_temp = dset_temp.sort_values(by = 'percentage_of_dot_balls',ascending = False).reset_index(drop=True)
dset_temp = dset_temp.iloc[:10,:]

fig,ax1 = plt.subplots(figsize=(15,8))
ax2 = ax1.twinx()
labels = np.array(dset_temp["batsman"])
ind = np.arange(len(labels))
width = 0.9
rects = ax1.bar(ind, np.array(dset_temp["dot_balls"]), width = width , color = 'brown')
ax1.set_xticks(ind+((width)/2.))
ax1.set_xticklabels(labels,rotation = "vertical")
ax1.set_ylabel("count of dot balls",color = "brown")
ax1.set_title("Batsman with highest percentage")
ax2.plot(ind+0.45, np.array (dset_temp["percentage_of_dot_balls"]),color = "b" ,marker = 'o')
ax2.set_ylabel("Percentage of dot balls", color = 'b')
ax2.set_ylim([0,100])
ax2.grid(b=False)
plt.show()
```

##### Batsman with more than 300 balls faced in taken and the ones with higher percentage of dot balls are seen. It is interesting to see Ganguly with more than 1000 balls and nearly half of them are dot balls. It is surprising to see names like Jayasuriya and Gibbs in there.!

##### Bowler Analysis:

Now let us see the bowlers who has bowled most number of balls in IPL.

```python
dset_temp = dset_score.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball', ascending=False).reset_index(drop=True)
dset_temp = dset_temp.iloc[:10,:]

labels = np.array(dset_temp['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(dset_temp['ball']), width=width, color='cyan')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of balls bowled in IPL")
ax.set_xlabel('Bowler Names')
autolabel(rects)
plt.show()
```

#### Harbhajan Singh is the the bowler with most number of balls bowled in IPL matches. Now let us see the bowler with more number of dot balls.

```python
dset_temp = dset_score.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs', ascending=False).reset_index(drop=True)
dset_temp = dset_temp.iloc[:10,:]

labels = np.array(dset_temp['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(dset_temp['total_runs']), width=width, color='lightskyblue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of dot balls bowled in IPL")
ax.set_xlabel('Bowler Names')
autolabel(rects)
plt.show()
```

##### Pravin Kumar is the one with more number of dot balls followed by Steyn and Malinga

```python
# Now let us see the bowlers who has bowled more number of extras in IPL.
dset_temp = dset_score.groupby('bowler')['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs', ascending=False).reset_index(drop=True)
dset_temp = dset_temp.iloc[:10,:]

labels = np.array(dset_temp['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(dset_temp['extra_runs']), width=width, color='magenta')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Bowlers with more extras in IPL")
ax.set_xlabel('Bowler Names')
autolabel(rects)
plt.show()
```

#### Malinga tops the chart with 221 extra runs followed by Pravin Kumar.
